## Part 2: Creating a Beautiful Board with Tables

Rich has a `Table` class perfect for our Mancala board:

````python
from rich.table import Table
from rich.console import Console

console = Console()

def create_board():
    """Create a new Mancala board."""
    return [4, 4, 4, 4, 4, 4, 0, 4, 4, 4, 4, 4, 4, 0]


def display_board_rich(board, player1_name="Player 1", player2_name="Player 2"):
    """
    Display the Mancala board using Rich tables.
    """
    # Create a table
    table = Table(title="🎮 MANCALA BOARD 🎮", 
                  title_style="bold cyan",
                  show_header=True,
                  header_style="bold magenta")
    
    # Add columns - Store, then 6 pockets, then Store
    table.add_column("Store", justify="center", style="yellow", width=8)
    for i in range(6):
        table.add_column(str(5-i), justify="center", style="cyan", width=6)
    table.add_column("Store", justify="center", style="yellow", width=8)
    
    # Player 2's row (top) - reversed order for pockets
    player2_row = [
        f"[bold yellow]{board[13]}[/bold yellow]",  # Player 2's store
        *[f"[cyan]{board[i]}[/cyan]" for i in range(12, 6, -1)],  # Pockets 12-7
        ""  # Empty cell
    ]
    table.add_row(*player2_row, style="on #2d2d2d")
    
    # Add separator row with pocket numbers
    separator = ["", "12", "11", "10", "9", "8", "7", ""]
    table.add_row(*separator, style="dim")
    
    # Player 1's row (bottom)
    player1_row = [
        "",  # Empty cell
        *[f"[green]{board[i]}[/green]" for i in range(0, 6)],  # Pockets 0-5
        f"[bold yellow]{board[6]}[/bold yellow]"  # Player 1's store
    ]
    table.add_row(*player1_row, style="on #2d2d2d")
    
    # Add row with pocket numbers
    pocket_nums = ["", "0", "1", "2", "3", "4", "5", ""]
    table.add_row(*pocket_nums, style="dim")
    
    # Display the table
    console.print()
    console.print(table)
    console.print(f"\n[bold green]{player1_name}[/bold green] vs [bold cyan]{player2_name}[/bold cyan]")
    console.print()


# Test it!
board = create_board()
display_board_rich(board, "Alice", "Bob")
````

### Understanding the Code

**Table creation:**
```python
table = Table(title="🎮 MANCALA BOARD 🎮")
```

**Adding columns:**
```python
table.add_column("Name", justify="center", style="cyan")
```

**Adding rows:**
```python
table.add_row("value1", "value2", "value3")
```

**Rich markup:**
- `[bold]text[/bold]` - Bold text
- `[red]text[/red]` - Red text
- `[bold yellow on blue]text[/bold yellow on blue]` - Yellow text on blue background

## Part 3: Adding Panels for Messages

Use panels to highlight important messages:

````python
from rich.panel import Panel

def show_welcome():
    """Display a welcome message."""
    welcome_text = """
    [bold cyan]Welcome to Mancala![/bold cyan]
    
    🎯 Goal: Collect the most seeds in your store
    ⭐ Land in your store for a free turn
    💰 Land in an empty pocket to capture!
    """
    
    panel = Panel(welcome_text, 
                  title="🎮 Game Rules 🎮",
                  border_style="green",
                  padding=(1, 2))
    
    console.print(panel)

# Test it
show_welcome()
````

## Part 4: Styled Move Announcements

Make move announcements more exciting:

````python
def announce_move(player, pocket, seeds):
    """Announce a player's move with style."""
    player_color = "green" if player == 1 else "cyan"
    
    console.print(
        f"\n[bold {player_color}]Player {player}[/bold {player_color}] "
        f"picks pocket [yellow]{pocket}[/yellow] "
        f"({seeds} 🌰 seeds)"
    )


def announce_capture(player, captured_seeds):
    """Announce a capture."""
    console.print(
        f"\n[bold red]💰 CAPTURE![/bold red] "
        f"[bold]Player {player}[/bold] captures "
        f"[yellow]{captured_seeds}[/yellow] seeds!"
    )


def announce_free_turn(player):
    """Announce a free turn."""
    console.print(
        f"\n[bold green]🎉 FREE TURN![/bold green] "
        f"Player {player} goes again!"
    )


# Test announcements
announce_move(1, 3, 4)
announce_capture(1, 8)
announce_free_turn(2)
````

## Part 5: Game Over Screen

Create an impressive game over display:

````python
from rich.table import Table
from rich.panel import Panel

def show_game_over(board, player1_name, player2_name):
    """Display game over screen with final scores."""
    player1_score = board[6]
    player2_score = board[13]
    
    # Create scores table
    scores_table = Table(show_header=True, header_style="bold magenta")
    scores_table.add_column("Player", style="cyan", justify="left")
    scores_table.add_column("Final Score", style="yellow", justify="center")
    
    scores_table.add_row(player1_name, str(player1_score))
    scores_table.add_row(player2_name, str(player2_score))
    
    # Determine winner
    if player1_score > player2_score:
        winner_text = f"[bold green]🏆 {player1_name} WINS! 🏆[/bold green]"
    elif player2_score > player1_score:
        winner_text = f"[bold cyan]🏆 {player2_name} WINS! 🏆[/bold cyan]"
    else:
        winner_text = "[bold yellow]🤝 TIE GAME! 🤝[/bold yellow]"
    
    # Create game over panel
    console.print()
    console.print(Panel.fit(
        "[bold red]GAME OVER[/bold red]",
        border_style="red"
    ))
    console.print()
    console.print(scores_table)
    console.print()
    console.print(Panel.fit(
        winner_text,
        border_style="gold1"
    ))
    console.print()


# Test game over screen
test_board = [0, 0, 0, 0, 0, 0, 28, 0, 0, 0, 0, 0, 0, 20]
show_game_over(test_board, "Alice", "Bob")
````

## Part 6: Adding Visual Seed Representations

Use Unicode symbols to show seeds visually:

````python
def display_board_with_seeds(board, player1_name="Player 1", player2_name="Player 2"):
    """
    Display board with visual seed representation.
    """
    def seed_display(count):
        """Show seeds visually with emojis."""
        if count == 0:
            return "[dim]empty[/dim]"
        elif count <= 6:
            return "🌰 " * count
        else:
            return f"🌰 × {count}"
    
    # Create main table
    table = Table(title="🎮 MANCALA BOARD 🎮",
                  title_style="bold cyan",
                  show_header=True,
                  header_style="bold magenta",
                  box=None)
    
    # Add columns
    table.add_column("P2 Store", justify="center", style="yellow", width=15)
    for i in range(6):
        table.add_column(str(12-i), justify="center", style="cyan", width=12)
    
    # Player 2's row with seed visualization
    p2_row = [
        f"[bold yellow]{board[13]}\n{seed_display(board[13])}[/bold yellow]"
    ]
    for i in range(12, 6, -1):
        p2_row.append(f"[cyan]{board[i]}\n{seed_display(board[i])}[/cyan]")
    
    table.add_row(*p2_row)
    
    # Separator
    table.add_row(*(["―" * 10] * 7), style="dim")
    
    # Add columns for Player 1's row
    table2 = Table(show_header=True, header_style="bold magenta", box=None)
    for i in range(6):
        table2.add_column(str(i), justify="center", style="green", width=12)
    table2.add_column("P1 Store", justify="center", style="yellow", width=15)
    
    # Player 1's row with seed visualization
    p1_row = []
    for i in range(0, 6):
        p1_row.append(f"[green]{board[i]}\n{seed_display(board[i])}[/green]")
    p1_row.append(f"[bold yellow]{board[6]}\n{seed_display(board[6])}[/bold yellow]")
    
    table2.add_row(*p1_row)
    
    # Display both tables
    console.print()
    console.print(table)
    console.print(table2)
    console.print(f"\n[bold green]{player1_name}[/bold green] vs [bold cyan]{player2_name}[/bold cyan]")
    console.print()


# Test visual seeds
test_board = [3, 0, 5, 2, 1, 4, 8, 4, 0, 3, 6, 2, 5, 10]
display_board_with_seeds(test_board, "Alice", "Bob")
````

## Part 7: Progress Indicator for AI Thinking

Add a spinner when AI is "thinking":

````python
from rich.spinner import Spinner
from rich.live import Live
import time

def ai_thinking_animation():
    """Show AI thinking with a spinner."""
    spinner = Spinner("dots", text="[cyan]AI is thinking...[/cyan]")
    
    with Live(spinner, refresh_per_second=10):
        time.sleep(2)  # Simulate AI thinking
    
    console.print("[green]✓[/green] AI has chosen a move!")


# Test it
ai_thinking_animation()
````

## Part 8: Statistics Dashboard

Create a nice statistics display:

````python
from rich.table import Table

def show_statistics(stats):
    """
    Display game statistics in a formatted table.
    
    Parameters:
        stats (dict): Dictionary with game statistics
    """
    table = Table(title="📊 Game Statistics", 
                  title_style="bold cyan",
                  show_header=True,
                  header_style="bold magenta")
    
    table.add_column("Statistic", style="cyan")
    table.add_column("Value", style="yellow", justify="right")
    
    table.add_row("Total Moves", str(stats.get('moves', 0)))
    table.add_row("Player 1 Moves", str(stats.get('player1_moves', 0)))
    table.add_row("Player 2 Moves", str(stats.get('player2_moves', 0)))
    table.add_row("Free Turns", str(stats.get('free_turns', 0)))
    table.add_row("Captures", str(stats.get('captures', 0)))
    
    console.print()
    console.print(table)
    console.print()


# Test statistics
test_stats = {
    'moves': 45,
    'player1_moves': 24,
    'player2_moves': 21,
    'free_turns': 8,
    'captures': 5
}
show_statistics(test_stats)
````
