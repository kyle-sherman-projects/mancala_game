## Part 1: Review - Setting Up Our Board

Let's bring back our code from Day 1:

````python
def create_board():
    """
    Create a new Mancala board with starting positions.
    """
    board = [4, 4, 4, 4, 4, 4, 0, 4, 4, 4, 4, 4, 4, 0]
    return board


def display_board(board):
    """
    Display the Mancala board in a readable format.
    """
    print("\n" + "=" * 50)
    print("                MANCALA BOARD")
    print("=" * 50)
    
    print(f"Player 2 Store: [{board[13]:2d}]")
    print(f"Player 2:  {board[12]:2d}  {board[11]:2d}  {board[10]:2d}  {board[9]:2d}  {board[8]:2d}  {board[7]:2d}")
    print("           " + "―" * 25)
    print(f"Player 1:  {board[0]:2d}  {board[1]:2d}  {board[2]:2d}  {board[3]:2d}  {board[4]:2d}  {board[5]:2d}")
    print(f"Player 1 Store: [{board[6]:2d}]")
    print("           0   1   2   3   4   5   (pocket numbers)")
    print("=" * 50 + "\n")


# Test it
board = create_board()
display_board(board)
````

## Part 2: Understanding the Move Logic

**How a Mancala move works:**

1. Player picks a pocket on their side
2. Player picks up ALL seeds from that pocket
3. Moving counter-clockwise, player drops one seed in each pocket/store
4. Player skips opponent's store
5. Special rules:
   - If last seed lands in your store → you get another turn
   - If last seed lands in your empty pocket → capture!

**Example:** Player 1 picks pocket 2 (which has 4 seeds)
```
Before:  [4, 4, 4, 4, 4, 4, 0, ...]
          0  1  2  3  4  5  6

After:   [4, 4, 0, 5, 5, 5, 1, ...]
          0  1  2  3  4  5  6
```

Seeds went: 2→3→4→5→6 (store)

## Part 3: Planning with Pseudocode

Before writing Python code, programmers write **pseudocode** - a plain English description of the algorithm.

**Pseudocode for a move:**
```
FUNCTION make_move(board, pocket, player):
    1. Check if pocket is valid (has seeds, on player's side)
    2. Pick up all seeds from chosen pocket
    3. Set that pocket to 0
    4. Starting from next position, drop one seed at a time
    5. Skip opponent's store
    6. Wrap around to beginning if needed
    7. Check for special rules (extra turn, capture)
    8. Return updated board
```

Now let's turn this into Python!

## Part 5: The Modulo Operator - Wrapping Around

When distributing seeds, we need to wrap around from position 13 back to 0.

The **modulo operator** `%` gives us the remainder after division:

````python
print("5 % 3 =", 5 % 3)   # 5 ÷ 3 = 1 remainder 2
print("10 % 4 =", 10 % 4) # 10 ÷ 4 = 2 remainder 2
print("14 % 14 =", 14 % 14) # 14 ÷ 14 = 1 remainder 0
print("15 % 14 =", 15 % 14) # 15 ÷ 14 = 1 remainder 1
````

**Why is this useful?** We can use `% 14` to wrap around our board:

```python
position = 13  # Last position on board
position = (position + 1) % 14  # Goes to 0 (wraps around!)
```

````python
# Demonstrate wrapping around the board
position = 11

for i in range(5):
    print(f"Current position: {position}")
    position = (position + 1) % 14

print(f"Final position: {position}")
````

## Part 6: Building the Move Function - Version 1 (Simple)

Let's start with a simple version that just distributes seeds (no special rules yet):

````python
def make_move_simple(board, pocket):
    """
    Make a simple move - pick up seeds and distribute them.
    No special rules yet (skipping opponent store, captures, etc.)
    
    Parameters:
    board (list): The current game board
    pocket (int): Which pocket to pick from (0-5 for Player 1)
    
    Returns:
    list: Updated board
    """
    # Pick up all seeds from chosen pocket
    seeds_in_hand = board[pocket]
    board[pocket] = 0  # Empty the pocket
    
    print(f"Picked up {seeds_in_hand} seeds from pocket {pocket}")
    
    # Start distributing from next position
    current_position = pocket
    
    # Drop one seed at a time
    while seeds_in_hand > 0:
        current_position = (current_position + 1) % 14  # Move to next spot
        board[current_position] += 1  # Add one seed
        seeds_in_hand -= 1  # One less seed in hand
        
        print(f"  Dropped seed at position {current_position}, {seeds_in_hand} seeds remaining")
    
    return board


# Test it!
print("TESTING SIMPLE MOVE\n")
test_board = create_board()
print("Before move:")
display_board(test_board)

print("\nPlayer 1 picks pocket 2:\n")
test_board = make_move_simple(test_board, 2)

print("\nAfter move:")
display_board(test_board)
````

## Part 7: Adding Player Logic and Skipping Opponent's Store

Now let's add:
- Player 1 vs Player 2 distinction
- Skipping the opponent's store

````python
def make_move(board, pocket, player):
    """
    Make a move with proper player logic.
    
    Parameters:
    board (list): The current game board
    pocket (int): Which pocket to pick from
    player (int): 1 or 2 (which player is moving)
    
    Returns:
    tuple: (updated board, last position where seed landed)
    """
    # Determine which store to skip
    if player == 1:
        opponent_store = 13
        player_store = 6
    else:
        opponent_store = 6
        player_store = 13
    
    # Pick up all seeds
    seeds_in_hand = board[pocket]
    board[pocket] = 0
    
    print(f"\nPlayer {player} picked up {seeds_in_hand} seeds from pocket {pocket}")
    
    current_position = pocket
    
    # Distribute seeds
    while seeds_in_hand > 0:
        current_position = (current_position + 1) % 14
        
        # Skip opponent's store
        if current_position == opponent_store:
            print(f"  Skipping opponent's store at position {opponent_store}")
            continue  # Skip this iteration, don't drop a seed
        
        board[current_position] += 1
        seeds_in_hand -= 1
        print(f"  Dropped seed at position {current_position}")
    
    print(f"\nLast seed landed at position {current_position}")
    
    return board, current_position


# Test with both players
print("=" * 50)
print("TESTING WITH PLAYER LOGIC")
print("=" * 50)

test_board = create_board()
display_board(test_board)

print("\n--- Player 1's turn ---")
test_board, last_pos = make_move(test_board, 0, player=1)
display_board(test_board)
````

## Part 9: Checking for Free Turn

**Rule:** If your last seed lands in your own store, you get another turn!

Let's add logic to detect this:

````python
def check_free_turn(last_position, player):
    """
    Check if player gets a free turn.
    
    Parameters:
    last_position (int): Where the last seed landed
    player (int): Which player made the move
    
    Returns:
    bool: True if player gets another turn
    """
    if player == 1 and last_position == 6:
        return True
    elif player == 2 and last_position == 13:
        return True
    else:
        return False


# Test it
print("Testing free turn detection:\n")

# This should give Player 1 a free turn
test_board = [2, 4, 4, 4, 4, 4, 0, 4, 4, 4, 4, 4, 4, 0]
display_board(test_board)

print("Player 1 picks pocket 0 (has 2 seeds):")
test_board, last_pos = make_move(test_board, 0, player=1)
display_board(test_board)

if check_free_turn(last_pos, player=1):
    print("🎉 Player 1 gets another turn!")
else:
    print("Turn switches to Player 2")
````

## Part 10: Capture Rule

**Rule:** If your last seed lands in one of YOUR empty pockets, you capture:
- That seed
- All seeds in the opposite pocket
- Both go to your store

**Opposite pocket mapping:**
```
Player 1's pocket 0 ↔ Player 2's pocket 12
Player 1's pocket 1 ↔ Player 2's pocket 11
Player 1's pocket 2 ↔ Player 2's pocket 10
Player 1's pocket 3 ↔ Player 2's pocket 9
Player 1's pocket 4 ↔ Player 2's pocket 8
Player 1's pocket 5 ↔ Player 2's pocket 7
```

Formula: `opposite = 12 - pocket`

````python
def get_opposite_pocket(pocket):
    """
    Get the pocket directly opposite on the board.
    
    Parameters:
    pocket (int): A pocket index (0-5 or 7-12)
    
    Returns:
    int: The opposite pocket index
    """
    return 12 - pocket


# Test opposite pocket calculation
print("Opposite pocket mappings:")
for pocket in range(6):
    opposite = get_opposite_pocket(pocket)
    print(f"Pocket {pocket} ↔ Pocket {opposite}")
````

````python
def check_capture(board, last_position, player):
    """
    Check if a capture occurred and execute it.
    
    Parameters:
    board (list): Current game board
    last_position (int): Where last seed landed
    player (int): Which player made the move
    
    Returns:
    list: Updated board (with capture if applicable)
    """
    # Capture only happens if:
    # 1. Last seed landed in player's own pocket (not store)
    # 2. That pocket now has exactly 1 seed (was empty before)
    # 3. Opposite pocket has seeds to capture
    
    player_pockets = range(0, 6) if player == 1 else range(7, 13)
    player_store = 6 if player == 1 else 13
    
    # Check if last position is in player's pockets
    if last_position not in player_pockets:
        return board  # No capture
    
    # Check if pocket now has exactly 1 seed (was empty)
    if board[last_position] != 1:
        return board  # No capture
    
    # Get opposite pocket
    opposite = get_opposite_pocket(last_position)
    
    # Check if opposite has seeds
    if board[opposite] == 0:
        return board  # No capture
    
    # CAPTURE!
    captured_seeds = board[opposite] + board[last_position]
    print(f"\n💰 CAPTURE! Player {player} captures {captured_seeds} seeds!")
    
    board[player_store] += captured_seeds
    board[opposite] = 0
    board[last_position] = 0
    
    return board


# Test capture scenario
print("TESTING CAPTURE RULE\n")
print("=" * 50)

# Set up a capture scenario
# Player 1's pocket 3 is empty, opposite pocket 9 has 5 seeds
capture_board = [4, 4, 1, 0, 4, 4, 0, 4, 4, 5, 4, 4, 4, 0]
print("Before move:")
display_board(capture_board)

print("\nPlayer 1 picks pocket 2 (will land in empty pocket 3):")
capture_board, last_pos = make_move(capture_board, 2, player=1)
print("\nAfter distributing seeds:")
display_board(capture_board)

print("\nChecking for capture...")
capture_board = check_capture(capture_board, last_pos, player=1)
print("\nFinal board after capture:")
display_board(capture_board)
````

## Part 11: Input Validation

We need to check if a move is legal before allowing it!

**A move is valid if:**
1. The pocket number is in the correct range
2. The pocket belongs to the current player
3. The pocket has seeds in it (not empty)

````python
def is_valid_move(board, pocket, player):
    """
    Check if a move is valid.
    
    Parameters:
    board (list): Current game board
    pocket (int): Pocket player wants to pick from
    player (int): Which player (1 or 2)
    
    Returns:
    tuple: (is_valid, error_message)
    """
    # Check 1: Is pocket number in valid range?
    if pocket < 0 or pocket > 12:
        return False, "Invalid pocket number. Choose 0-5."
    
    # Check 2: Is pocket on player's side?
    if player == 1:
        if pocket < 0 or pocket > 5:
            return False, "That pocket is not on your side. Choose 0-5."
    else:  # Player 2
        if pocket < 7 or pocket > 12:
            return False, "That pocket is not on your side. Choose 7-12."
    
    # Check 3: Does pocket have seeds?
    if board[pocket] == 0:
        return False, "That pocket is empty. Choose a different pocket."
    
    # All checks passed!
    return True, "Valid move!"


# Test validation
print("TESTING MOVE VALIDATION\n")
test_board = create_board()

# Test various scenarios
test_cases = [
    (2, 1, "Valid move for Player 1"),
    (8, 2, "Valid move for Player 2"),
    (10, 1, "Invalid - wrong side"),
    (2, 2, "Invalid - wrong side"),
]

for pocket, player, description in test_cases:
    valid, message = is_valid_move(test_board, pocket, player)
    status = "✓" if valid else "✗"
    print(f"{status} {description}")
    print(f"   Pocket {pocket}, Player {player}: {message}\n")

# Test with empty pocket
empty_board = [0, 4, 4, 4, 4, 4, 0, 4, 4, 4, 4, 4, 4, 0]
valid, message = is_valid_move(empty_board, 0, 1)
print(f"Empty pocket test: {message}")
````

## Part 12: Complete Move Function with All Rules

Let's put it all together into one comprehensive function:

````python
def make_complete_move(board, pocket, player, show_details=True):
    """
    Make a complete move with all Mancala rules.
    
    Parameters:
    board (list): Current game board
    pocket (int): Pocket to pick from
    player (int): Current player (1 or 2)
    show_details (bool): Whether to print move details
    
    Returns:
    tuple: (updated_board, gets_free_turn)
    """
    # Validate move
    valid, message = is_valid_move(board, pocket, player)
    if not valid:
        print(f"❌ Invalid move: {message}")
        return board, False
    
    # Determine stores
    if player == 1:
        opponent_store = 13
        player_store = 6
    else:
        opponent_store = 6
        player_store = 13
    
    # Pick up seeds
    seeds_in_hand = board[pocket]
    board[pocket] = 0
    
    if show_details:
        print(f"\nPlayer {player} picked up {seeds_in_hand} seeds from pocket {pocket}")
    
    current_position = pocket
    
    # Distribute seeds
    while seeds_in_hand > 0:
        current_position = (current_position + 1) % 14
        
        # Skip opponent's store
        if current_position == opponent_store:
            continue
        
        board[current_position] += 1
        seeds_in_hand -= 1
    
    if show_details:
        print(f"Last seed landed at position {current_position}")
    
    # Check for capture
    board = check_capture(board, current_position, player)
    
    # Check for free turn
    gets_free_turn = check_free_turn(current_position, player)
    
    if gets_free_turn and show_details:
        print(f"\n🎉 Player {player} gets another turn!")
    
    return board, gets_free_turn


# Full test game sequence
print("=" * 50)
print("COMPLETE MOVE FUNCTION TEST")
print("=" * 50)

game_board = create_board()
display_board(game_board)

# Move 1: Player 1
print("\n--- MOVE 1 ---")
game_board, free_turn = make_complete_move(game_board, 2, player=1)
display_board(game_board)

# Move 2: Player 2
print("\n--- MOVE 2 ---")
game_board, free_turn = make_complete_move(game_board, 9, player=2)
display_board(game_board)
````
