## Part 1: Understanding Mancala

Before we code, let's understand the game:

**Mancala Board Layout:**
```
        Player 2's Side
    [13] 12  11  10  9   8  [7]
    [0]  1   2   3   4   5  [6]
        Player 1's Side
```

- Each player has 6 pockets (small holes) with seeds
- Each player has 1 store (large pocket on their side)
- Player 1's store is index 6, Player 2's store is index 13
- Game starts with 4 seeds in each pocket

## Part 5: Creating the Mancala Board

Now let's build our actual game board! Remember:
- 14 total positions (6 pockets + 1 store for each player)
- Pockets start with 4 seeds each
- Stores start with 0 seeds

**Index mapping:**
```
Player 2: [13] 12 11 10  9  8 [7]
Player 1: [0]   1  2  3  4  5 [6]
```

````python
# Create the initial Mancala board
# Indices 0-5: Player 1's pockets
# Index 6: Player 1's store
# Indices 7-12: Player 2's pockets
# Index 13: Player 2's store

board = [4, 4, 4, 4, 4, 4, 0, 4, 4, 4, 4, 4, 4, 0]

print("Starting board:", board)
print("Player 1's store (index 6):", board[6])
print("Player 2's store (index 13):", board[13])
````

**Challenge Question:** Why does Player 1's store have index 6, not 7?

<details>
<summary>Click to reveal answer</summary>
Because Python starts counting at 0! If we have positions 0-5 (6 pockets), the next position is 6.
</details>

## Part 6: Displaying the Board

A list of numbers isn't easy to read. Let's make it look like a real Mancala board!

We want it to look like this:
```
Player 2:  4  4  4  4  4  4  | Store: 0
Player 1:  4  4  4  4  4  4  | Store: 0
           0  1  2  3  4  5
```

````python
def display_board(board):
    """
    Display the Mancala board in a readable format.
    
    Parameters:
    board (list): A list of 14 numbers representing the game state
    """
    # Player 2's side (indices 12 to 7, going right to left)
    print("\n" + "=" * 50)
    print("                MANCALA BOARD")
    print("=" * 50)
    
    # Player 2's store and pockets
    print(f"Player 2 Store: [{board[13]:2d}]")
    print(f"Player 2:  {board[12]:2d}  {board[11]:2d}  {board[10]:2d}  {board[9]:2d}  {board[8]:2d}  {board[7]:2d}")
    print("           " + "―" * 25)
    
    # Player 1's pockets and store
    print(f"Player 1:  {board[0]:2d}  {board[1]:2d}  {board[2]:2d}  {board[3]:2d}  {board[4]:2d}  {board[5]:2d}")
    print(f"Player 1 Store: [{board[6]:2d}]")
    print("           0   1   2   3   4   5   (pocket numbers)")
    print("=" * 50 + "\n")

# Test our display function
board = [4, 4, 4, 4, 4, 4, 0, 4, 4, 4, 4, 4, 4, 0]
display_board(board)
````

## Part 9: Putting It All Together

Let's create a complete starting template for our game:

````python
# ========================================
# MANCALA GAME - DAY 1 COMPLETE CODE
# ========================================

def create_board():
    """
    Create a new Mancala board with starting positions.
    Returns a list of 14 integers.
    """
    # 4 seeds in each pocket (0-5 and 7-12)
    # 0 seeds in each store (6 and 13)
    board = [4, 4, 4, 4, 4, 4, 0, 4, 4, 4, 4, 4, 4, 0]
    return board


def display_board(board):
    """
    Display the Mancala board in a readable format.
    """
    print("\n" + "=" * 50)
    print("                MANCALA BOARD")
    print("=" * 50)
    
    print(f"Player 2 Store: [{board[13]:2d}]")
    print(f"Player 2:  {board[12]:2d}  {board[11]:2d}  {board[10]:2d}  {board[9]:2d}  {board[8]:2d}  {board[7]:2d}")
    print("           " + "―" * 25)
    print(f"Player 1:  {board[0]:2d}  {board[1]:2d}  {board[2]:2d}  {board[3]:2d}  {board[4]:2d}  {board[5]:2d}")
    print(f"Player 1 Store: [{board[6]:2d}]")
    print("           0   1   2   3   4   5   (pocket numbers)")
    print("=" * 50 + "\n")


# Main game initialization
print("Welcome to Mancala!")
print("Let's see the starting board:\n")

game_board = create_board()
display_board(game_board)

print("Day 1 Complete! Tomorrow we'll make moves happen.")
````
